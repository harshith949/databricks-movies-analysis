```python
%python
from pyspark.sql import functions as F

df = spark.table("movies")

industry_metrics = (
    df
    .withColumn(
        "revenue_clean",
        F.when(
            F.regexp_replace("revenue", "[^0-9.]", "").cast("double").isNotNull(),
            F.regexp_replace("revenue", "[^0-9.]", "").cast("double")
        ).otherwise(0.0)
    )
    .withColumn(
        "budget_clean",
        F.when(
            F.regexp_replace("budget", "[^0-9.]", "").cast("double").isNotNull(),
            F.regexp_replace("budget", "[^0-9.]", "").cast("double")
        ).otherwise(0.0)
    )
    .groupBy("industry")
    .agg(
        F.round(F.sum("revenue_clean"), 2).alias("total_revenue"),
        F.round(F.sum("budget_clean"), 2).alias("total_budget")
    )
    .orderBy(F.desc("total_revenue"))
)

display(industry_metrics)
```

```python
display(industry_metrics)
```

```python
pdf = industry_metrics.toPandas()

import matplotlib.pyplot as plt

plt.figure()
plt.bar(pdf["industry"], pdf["total_revenue"])
plt.xticks(rotation=45)
plt.title("Revenue by Industry")
plt.xlabel("Industry")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()
```
